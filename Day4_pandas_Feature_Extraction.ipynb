```python
def extract_episode(txt):
  check = False
  data = ''
  for i in txt:
     if i == ')' :
      check = False
      return data
     if check == True :
      data += i
     if i == '(' :
      check = True




```

```python
df['episode']=df['Title'].apply(extract_episode)
```

```python
def extract_timestamp(txt):
  check = False;
  time_stamp = ""
  count = 0
  for i in txt:
    if  (count<=18) and (check == True) :
      time_stamp += i
      count +=1
    if count > 18 :
      return time_stamp
    if i == ')':
      check = True
  return time_stamp
```

```python
df['Timestamp']=df['Title'].apply(extract_timestamp)
```

```python
def calculate_months(timestamp):
    if not isinstance(timestamp, str):
        return None

    month_map = {
        "Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4,
        "May": 5, "Jun": 6, "Jul": 7, "Aug": 8,
        "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12
    }

    try:
        start, end = timestamp.split('-')

        start_month, start_year = start.strip().split()
        end_month, end_year = end.strip().split()

        start_total = int(start_year) * 12 + month_map[start_month]
        end_total = int(end_year) * 12 + month_map[end_month]

        # +1 to count both start and end months
        return end_total - start_total + 1

    except Exception:
        return None
```

```python
df['Months'] = df['Timestamp'].apply(calculate_months)
df.head()
```
