#Mathematical Operations

1. Basic Arithmetic Operations

```python
#All Operations are perfomed element wise
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a+b,"\n")
print(a-b,"\n")
print(a*b,"\n")
print(a/b,"\n")
print(a**b,"\n")
print(a%b)
```

2.Universal Mathematical Functions

```python

print(np.sqrt(a))


print(np.exp(a))


print(np.log(a))


print(np.sin(a))   # [0.84147098 0.90929743 0.14112001]
print(np.cos(a))   # [0.54030231 -0.41614684 -0.9899925]
```

3. Aggregation Functions

```python
arr = np.array([1, 2, 3, 4, 5])

print(np.sum(arr))       # 15
print(np.mean(arr))      # 3.0
print(np.std(arr))       # 1.414213562
print(np.min(arr))       # 1
print(np.max(arr))       # 5
```

```python
matrix = np.array([[1,2,3],[4,5,6]])
print(np.sum(matrix, axis=0))  # sum along columns: [5 7 9]
print(np.sum(matrix, axis=1))  # sum along rows: [6 15]
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

# Matrix multiplication
print(np.dot(A, B))
# [[19 22]
#  [43 50]]

# Transpose
print(A.T)
# [[1 3]
#  [2 4]]

# Determinant
print(np.linalg.det(A))  # -2.0
```
